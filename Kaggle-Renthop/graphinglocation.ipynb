```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_location = pd.read_csv('data/train_location.csv')
```

```python
x = set(train_location.neighborhood_cleaned)
cols = {} 
n = 1
for i in x:
    cols[i] = n
    n += 1
```

```python
train_location.neighborhood_cleaned = train_location.neighborhood_cleaned.apply(lambda x: cols[x])
```

```python
train_location.head
```

```python
plt.xlim([-74.05,-73.85])
plt.ylim([40.60, 40.90])
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.title('Neighborhoods')
plt.scatter(train_location.longitude, train_location.latitude, \
            c = train_location.neighborhood_cleaned, cmap=plt.cm.rainbow)
plt.savefig('neighborhoods_plot.png')
```
